```python
#checking the unique values in the ship mode column
df['Ship Mode'].unique()
#changing the ambiguous value to nan
df=pd.read_csv('orders.csv', na_values=['Not Available', 'unknown'])
df.head(10)
```

```python
#fixing the columns name
df.columns=df.columns.str.lower()
df.columns= df.columns.str.replace(' ','_')
df.head()
```

```python
#deriving new columns discount,sales price and profit
df['discount']=df['list_price']*df['discount_percent']*0.01
df['sales_price']=df['list_price']-df['discount']
df['profit']=df['cost_price']-df['sales_price']
```

```python
#dropping unrequired columns
df.drop(columns=['cost_price','list_price','discount_percent'],inplace=True)
df.head()
```

```python
#converting the order_date in date time format
df['order_date']=pd.to_datetime(df['order_date'],format="%Y-%m-%d")
```
